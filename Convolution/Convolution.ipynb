# Bibliotecas e funções

```python
import numpy as np
import matplotlib.pyplot as plt

from Convolution import *
```

## Plot de figuras

```python
def show_results(signal1, signal2, convolved_signal):

  plt.figure(figsize=(16, 4))

  # Plot do sinal 1
  plt.subplot(1,2,1)
  plt.stem(signal1)
  plt.title('Primeiro Sinal')
  plt.xlabel('n')
  plt.ylabel('Amplitude')
  plt.grid(True)

  # Plot do sinal 2
  plt.subplot(1,2,2)
  plt.stem(signal2)
  plt.title('Segundo Sinal')
  plt.xlabel('n')
  plt.ylabel('Amplitude')
  plt.grid(True)

  # Plot da convolução do sinal 1 e 2
  plt.figure(figsize=(16, 4))
  plt.stem(convolved_signal)
  plt.title('Resultado da convolução')
  plt.xlabel('n')
  plt.ylabel('Amplitude')
  plt.grid(True)

  plt.show()
```

# Demostração

### Cosseno e impulso

```python
x_n = np.cos(2*np.pi/20*np.arange(1, 20))
h_n = np.ones(1)
y_n = myConvolution(x_n, h_n)

show_results(x_n, h_n, y_n)
```

### Dois Degraus de mesmo tamanho

```python
x_n = np.ones(4)
h_n = np.ones(4)
y_n = myConvolution(x_n, h_n)

show_results(x_n, h_n, y_n)
```

### Dois Degraus de tamanho diferentes

```python
x_n = np.ones(4)
h_n = np.ones(7)
y_n = myConvolution(x_n, h_n)

show_results(x_n, h_n, y_n)
```

### Triangular e trem de impulsos


```python
x1_n = np.ones(3)
x2_n = np.ones(3)

x_n = myConvolution(x1_n, x2_n)
h_n = np.zeros(22)

for i in range(4):
  h_n[i*7] = 1

y_n = myConvolution(x_n, h_n)

show_results(x_n, h_n, y_n)
```
