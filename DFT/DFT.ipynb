# Bibliotecas utilizadas

```python
import numpy as np
import matplotlib.pyplot as plt

from DFT import *
```

# Função para mostrar os resultados

```python
def show_results(x_n, N=0):
    if not N:
        N = int(2**np.ceil(np.log2(len(x_n))))

    x_np = zero_padding(x_n, N)
    X_k = calculate_DFT(x_np)

    X_k_mod = np.abs(X_k)
    X_k_phase = np.angle(X_k)

    #===========================================================#
    plt.figure(figsize=(20, 6))
    plt.subplot(1, 2, 1)
    plt.stem(x_n)
    plt.title('Sinal de entrada x[n]')
    plt.xlabel('Amostras')
    plt.ylabel('Amplitude')
    plt.grid()

    plt.subplot(1, 2, 2)
    plt.stem(x_np)
    plt.title('Sinal de entrada x[n] com zero padding')
    plt.xlabel('Amostras')
    plt.ylabel('Magnitude')
    plt.grid()

    #===========================================================#

    plt.figure(figsize=(20, 6))
    plt.subplot(1, 2, 1)
    plt.stem(X_k_mod)
    plt.title('Magnitude da DFT')
    plt.xlabel('Amostras')
    plt.ylabel('Magnitude')
    plt.grid()

    plt.subplot(1, 2, 2)
    plt.stem(X_k_phase)
    plt.title('Fase of DFT')
    plt.xlabel('Amostras')
    plt.ylabel('Fase (rad)')
    plt.grid()

    return
```

# Exemplos

## Função Impulso na origem

```python
x_n = np.zeros(4)
x_n[0] = 1  # Impulso na origem

show_results(x_n, N=32)
```

## Função Impulso deslocado

```python
x_n = np.zeros(4)
x_n[2] = 1  # Impulso na origem

show_results(x_n, N=32)
```

## Função Cosseno

```python
x_n = np.cos(np.linspace(0, 2 * np.pi, 32))  # Example signal
show_results(x_n)
```
